```python
dfRekap = pd.read_csv('mixedRekap.csv')
dfRekap.drop([0, 2,4,6,8,10,12,14,16])
```

```python
gps = dfRekap['distGps'].mean()
noKal = dfRekap['distNoKal'].mean()
kal = dfRekap['distKal'].mean()
print('Mean distance error on gps: ', str(gps), ', while mean distance error without filter: ', str(noKal),
      ', and distance error with filter: ', str(kal))
```

```python
nAcuanLat1 = dfRekap['latMyGps']
nTerukurLat1 = dfRekap['latGps']

# nAcuan = 697645
# nTerukur = 697646
errorRateLat1 = (((nAcuanLat1-nTerukurLat1)/nAcuanLat1)*100)
print('Error rate LatGps')
print(errorRateLat1)
nAcuanLong1 = dfRekap['longMyGps']
nTerukurLong1 = dfRekap['longGps']

# nAcuan = 697645
# nTerukur = 697646
errorRateLong1 = (((nAcuanLong1-nTerukurLong1)/nAcuanLong1)*100)
print('Error rate longGps')
print(errorRateLong1)
```
